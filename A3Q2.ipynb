```python
import networkx as nx
```

```python
# Load the dataset and create a directed graph
data = nx.read_edgelist('/content/p2p-Gnutella09.txt', create_using=nx.DiGraph())
```

```python
# Compute PageRank score for each node
pagerank_scores = nx.pagerank(data)
# Compute Authority and Hub scores for each node using HITS algorithm
hits_scores = nx.hits(data)
```

```python
# Print the scores for each node
for node in data.nodes():
  print('Node:', node)
  print('PageRank Score:', pagerank_scores[node] )
  print('---------------------------------')
```

```python
for node in data.nodes():
  print('Node:', node)
  print('Authority Score:', hits_scores[0][node])
  print('Hub Score:', hits_scores[1][node])
  print('---------------------------------')
```

```python
import matplotlib.pyplot as plt

# Select the first 100 nodes
nodes = list(data.nodes())[:50]

# Compute PageRank score for each node
pagerank_scores = nx.pagerank(data)

# Compute Authority and Hub scores for each node using HITS algorithm
hits_scores = nx.hits(data)
authority_scores = hits_scores[0]
hub_scores = hits_scores[1]

# Create lists of node IDs and their corresponding scores
pagerank_scores_list = [pagerank_scores[node] for node in nodes]
authority_scores_list = [authority_scores[node] for node in nodes]
hub_scores_list = [hub_scores[node] for node in nodes]

# Create a bar chart of the scores
fig, ax = plt.subplots(figsize=(70, 70))
ax.bar(nodes, pagerank_scores_list, label="PageRank Score")
ax.bar(nodes, authority_scores_list, label="Authority Score")
ax.bar(nodes, hub_scores_list, label="Hub Score")
ax.legend()
ax.set_xlabel("Nodes",fontsize=50)
ax.set_ylabel("Scores",fontsize=50)
ax.set_title("Scores of nodes in the graph")
plt.show()
```
