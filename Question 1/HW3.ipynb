```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
```

```python
file_path = "/content/p2p-Gnutella09.txt"
```

```python

# Load the data from the txt file
with open(file_path, 'r') as f:
    data = f.read().splitlines()[4:]

# Create a dictionary to store the nodes and their connections
nodes = {}
for line in data:
    from_node, to_node = line.split()
    if from_node not in nodes:
        nodes[from_node] = []
    nodes[from_node].append(to_node)

# Get the number of nodes
num_nodes = len(nodes)

# Create an empty adjacency matrix
adj_matrix = [[0 for _ in range(num_nodes)] for _ in range(num_nodes)]

# Fill the adjacency matrix with 1s and 0s
for i in range(num_nodes):
    for j in range(num_nodes):
        if str(j) in nodes.get(str(i), []):
            adj_matrix[i][j] = 1



# Print the adjacency matrix
print("Adjacency matrix:")
for row in adj_matrix:
    print(row)
```

```python
# Create the edge list
edge_list = []
for node, connections in nodes.items():
    for connection in connections:
        edge_list.append((node, connection))

# Print the edge list
print("\nEdge list:")
for edge in edge_list:
    print(edge[0], "->", edge[1])
```

```python
with open(file_path, 'r') as f:
    lines = f.readlines()[4:]
    edges = []
    nodes = set()
    for line in lines:
        source, target = line.strip().split('\t')
        edges.append((source, target))
        nodes.add(source)
        nodes.add(target)

# Number of nodes
num_nodes = len(nodes)

# Number of edges
num_edges = len(edges)

# In-degree and out-degree dictionaries for each node
in_degrees = {}
out_degrees = {}
for node in nodes:
    in_degrees[node] = 0
    out_degrees[node] = 0

# Calculate in-degree and out-degree for each node
for edge in edges:
    source = edge[0]
    target = edge[1]
    out_degrees[source] += 1
    in_degrees[target] += 1

# Average in-degree and out-degree
avg_in_degree = sum(in_degrees.values()) / num_nodes
avg_out_degree = sum(out_degrees.values()) / num_nodes

# Node with max in-degree and max out-degree
max_in_degree_node = max(in_degrees, key=in_degrees.get)
max_out_degree_node = max(out_degrees, key=out_degrees.get)

# Network density
density = num_edges / (num_nodes * (num_nodes - 1))

print("Number of Nodes:", num_nodes)
print("Number of Edges:", num_edges)
print("Avg In-degree:", avg_in_degree)
print("Avg Out-degree:", avg_out_degree)
print("Node with Max In-degree:", max_in_degree_node)
print("Node with Max out-degree:", max_out_degree_node)
print("Density of the network:", density)
```

```python
# Load the dataset
data = np.loadtxt(file_path)
nodes = []
for edge in data:
    nodes.append(int(edge[0]))
    nodes.append(int(edge[1]))

# Create an empty dictionary to store the in-degree of each node
in_degree = {}

# Calculate the in-degree of each node
for edge in data:
    node = int(edge[1])
    if node not in in_degree:
        in_degree[node] = 1
    else:
        in_degree[node] += 1

out_degree = {}

# Calculate the out-degree of each node
for edge in data:
    node = int(edge[0])
    if node not in out_degree:
        out_degree[node] = 1
    else:
        out_degree[node] += 1

# nodes which do not have in-degree or out-degree   
for i in range (0 , 7115):
    if i not in in_degree.keys():
        in_degree[i] = 0
    if i not in out_degree.keys():
        out_degree[i] = 0


plt.figure(figsize=(15,10))
plt.hist(in_degree.values(), bins=80, color='green')
plt.title('In-degree')
plt.xlabel('Degree')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(15,10))
plt.hist(out_degree.values(), bins=90, color='orange')
plt.title('Out-degree')
plt.xlabel('Degree')
plt.ylabel('Count')
plt.show()
```

```python
# Load the dataset
content = np.loadtxt(file_path)

# Find the number of nodes in the network
num_nodes = int(max(np.max(content[:, 0]), np.max(content[:, 1]))) + 1

# Create an empty dictionary to store the neighbors of each node
neighbors = {i: set() for i in range(num_nodes)}

# Populate the dictionary with the neighbors of each node
for edge in content:
    neighbors[int(edge[0])].add(int(edge[1]))
    neighbors[int(edge[1])].add(int(edge[0]))

# Create an empty dictionary to store the local clustering coefficient of each node
lcc = {}

# Calculate the local clustering coefficient of each node
for node in range(num_nodes):
    neighbors_of_node = neighbors[node]
    num_neighbors = len(neighbors_of_node)
    if num_neighbors < 2:
        lcc[node] = 0.0
    else:
        num_edges_between_neighbors = 0
        for neighbor1 in neighbors_of_node:  
            for neighbor2 in neighbors_of_node:
                
                if neighbor1 < neighbor2 and neighbor2 in neighbors[neighbor1]:
                    num_edges_between_neighbors += 1
        max_possible_edges_between_neighbors = (num_neighbors * (num_neighbors - 1)) / 2
        lcc[node] = num_edges_between_neighbors / max_possible_edges_between_neighbors


# Plot the clustering-coefficient distribution

lcc_freq = defaultdict(int)
for node, cc in lcc.items():
    lcc_freq[cc] += 1

lcc_values = sorted(lcc_freq.keys())
lcc_counts = [lcc_freq[lcc_val] for lcc_val in lcc_values]

plt.figure(figsize=(15,12))
plt.bar(lcc_values, lcc_counts, width=0.01)
plt.xlabel('Local Clustering Coefficient')
plt.ylabel('Frequency')
plt.title('Clustering Coefficient Distribution')
plt.yticks(range(0, 8000, 200))
plt.show()
```
